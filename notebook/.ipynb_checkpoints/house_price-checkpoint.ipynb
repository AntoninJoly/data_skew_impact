```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import os
import sys
sys.path.append('../src')
from skew import *

import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_squared_log_error
import pickle

data_dir = '../data/house-prices-advanced-regression-techniques'
```

```python
df = pd.read_csv(os.path.join(data_dir,'train.csv'))
df.head()
```

```python
target = 'SalePrice'
mockup = SkewAnalysis()
nan_drop, _ = mockup.handle_nan(df, target_col=target)
df_unskewed, skew_dict = mockup.unskew_correlated_variables()
df_proc, df_continuous, df_categorical = mockup.normalize_and_encode(keep_corr = True)

df_proc.head()
```

# Training

```python
X_train, X, y_train, y = train_test_split(df_proc, df_proc[target], test_size = 0.2, random_state=42, shuffle=True)
X_val, X_test, y_val, y_test = train_test_split(X, y, test_size = 0.5, random_state=42, shuffle=True)
```

```python
# Hyperparameters of xgb regression algorithm (optimization can be studied in a different notebook)
params_xgb = {'eval_metric': 'rmse',
              'subsample': 0.6,
              'learning_rate': 0.01,
              'min_child_weight': 5,
              'colsample_bytree': 0.7,
              'objective': 'reg:squarederror',
              'n_estimators': 1000,
              'max_depth': 5}

# XGBoost setting and training
reg = xgb.XGBRegressor(**params_xgb).fit(X_train,
                                         y_train,
                                         early_stopping_rounds = 25,
                                         eval_metric = ['rmse'],
                                         eval_set = [(X_train, y_train), (X_val, y_val)],
                                         verbose = False)
```

```python
# Training metrics
results = reg.evals_result()
epochs = len(results['validation_0']['rmse'])
x_axis = range(0, epochs)

# Plot log loss
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,5))
ax1.plot(x_axis, results['validation_0']['rmse'], label='Train')
ax1.plot(x_axis, results['validation_1']['rmse'], label='Val')
ax1.legend(fontsize=15)
ax1.set_xlabel('epoch', fontsize=15)
ax1.set_ylabel('Log Loss', fontsize=15)

plt.tight_layout()
```

```python
preds, gt, index = mockup.decode(reg.predict(X_test), y_test)
```

```python
# Predict target values
plt.figure(figsize=(10,7))

# Root mean squared logaritmic error - Mean absolute error
# rmsle = np.sqrt(mean_squared_log_error(y_test, preds))
rmse = np.sqrt(mean_squared_error(preds, gt))

# Plotting results
plt.plot(gt,'b-', label='Ground truth')
plt.plot(preds,'r-', label='Prediction')
plt.ylabel('Price prediction')
plt.title('RMSE = %s'%(round(rmse, 4)))
plt.legend()
plt.show()
```
